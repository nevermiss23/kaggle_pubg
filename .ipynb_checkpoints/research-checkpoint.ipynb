# Generating features

```python
# Overall distance which player has overcame
data_tr['overall_distance'] = data_tr.ridedistance.values+data_tr.swimdistance.values+data_tr.walkdistance.values

sns.distplot(data_tr.overall_distance);
```

```python
data_tr[['overall_distance', 'winplaceperc']].corr()
```

```python
# Mean number of kills in a killstreak
data_tr['mean_klls_kllstr'] = data_tr.kills.values/data_tr.killstreaks.values
data_tr['mean_klls_kllstr'] = data_tr['mean_klls_kllstr'].fillna(0)

sns.distplot(data_tr.mean_klls_kllstr);
```

```python
# Number of items used
data_tr['items_used'] = data_tr.boosts.values+data_tr.heals.values

sns.distplot(data_tr.mean_klls_kllstr);
```
